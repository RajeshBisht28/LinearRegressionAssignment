### Eliminate Outliers

### major Outliers found Sale and Weight : So remove: Sale column, and remove outlier of Weight range

```python
q1 = df['weight'].quantile(0.25)
q3 = df['weight'].quantile(0.75)

iqr = q3 - q1
lower_bound = q1 - (1.5 * iqr)
upper_bound = q3 + (1.5 * iqr)

outlier = (df['weight'] < lower_bound) | (df['weight'] > upper_bound)
df = df[~outlier]
```

```python
df.drop(columns=['Product_id','Sale'],inplace=True)
```

```python
df.shape
```

```python
q1 = df['battery'].quantile(0.25)
q3 = df['battery'].quantile(0.75)

iqr = q3 - q1
lower_bound = q1 - (1.5 * iqr)
upper_bound = q3 + (1.5 * iqr)

outlier = (df['battery'] < lower_bound) | (df['battery'] > upper_bound)
total_outliers = np.sum(outlier)
print(f"Total Outliers of Battery: {total_outliers}")
```

```python
total_rows = len(df)
outliers_percentage = (total_outliers / total_rows) * 100
print(f"Total % of Battery outliers:{outliers_percentage}")
```

### Separated Dependent and Independent variable

```python
X = df.drop("Price", axis = 1)
y = df["Price"]
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Import Warnings
import warnings
warnings.filterwarnings('ignore')

# Sklearn Imports
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import StandardScaler,PolynomialFeatures,PowerTransformer
from sklearn.model_selection import cross_val_score
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline

from sklearn.metrics import mean_squared_error,r2_score

# Feature Selection
####from mlxtend.feature_selection import SequentialFeatureSelector
```

```python
## Training Testing Split : # Split into training (80%) and testing (20%) dat
X_train,X_test,y_train,y_test = train_test_split(X, y, test_size = 0.1, random_state = 101)
```

```python
model = LinearRegression()
model.fit(X_train, y_train)
# Predict on both the training and test sets
y_train_pred = model.predict(X_train)
y_test_pred = model.predict(X_test)

# Calculate the Mean Squared Error (MSE) for both training and test sets
train_mse = mean_squared_error(y_train, y_train_pred)
test_mse = mean_squared_error(y_test, y_test_pred)
train_r2 = r2_score(y_train, y_train_pred)
test_r2 = r2_score(y_test, y_test_pred)
print(f"Training MSE: {train_mse}")
print(f"Testing MSE: {test_mse}")
print(f"Training R2: {train_r2}")
print(f"Testing R2: {test_r2}")
```

### The difference between the training and testing MSE is relatively small. This suggests that the model is not overfitting. 

### Create and train Ridge Regression model

```python
from sklearn.linear_model import Ridge

ridge_model = Ridge(alpha=1.0)  # alpha is the regularization strength (higher value = more regularization)
ridge_model.fit(X_train, y_train)

# Predict on training and test sets
ridge_train_pred = ridge_model.predict(X_train)
ridge_test_pred = ridge_model.predict(X_test)

# Evaluate performance
ridge_train_mse = mean_squared_error(y_train, ridge_train_pred)
ridge_test_mse = mean_squared_error(y_test, ridge_test_pred)
train_ridge_r2 = r2_score(y_train, ridge_train_pred)
test_ridge_r2 = r2_score(y_test, ridge_test_pred)
print(f"Ridge Training MSE: {ridge_train_mse}")
print(f"Ridge Testing MSE: {ridge_test_mse}")
print(f"Ridge R2: {train_ridge_r2}")
print(f"Ridge R2: {test_ridge_r2}")
```

### Lasso Regression (L1 Regularization)

```python
from sklearn.linear_model import Lasso

# Create and train Lasso Regression model
lasso_model = Lasso(alpha=0.01)  # alpha is the regularization strength (higher value = more regularization)
lasso_model.fit(X_train, y_train)

# Predict on training and test sets
lasso_train_pred = lasso_model.predict(X_train)
lasso_test_pred = lasso_model.predict(X_test)

# Evaluate performance
lasso_train_mse = mean_squared_error(y_train, lasso_train_pred)
lasso_test_mse = mean_squared_error(y_test, lasso_test_pred)
train_lasso_r2 = r2_score(y_train, lasso_train_pred)
test_lasso_r2 = r2_score(y_test, lasso_test_pred)
print(f"Lasso Training MSE: {lasso_train_mse}")
print(f"Lasso Testing MSE: {lasso_test_mse}")
print(f"Lasso Training R2: {train_lasso_r2}")
print(f"Lasso Testing R2: {test_lasso_r2}")
```

####  ElasticNet (Combination of L1 and L2 Regularization)

```python
from sklearn.linear_model import ElasticNet

# Create and train ElasticNet model
elastic_net_model = ElasticNet(alpha=0.01, l1_ratio=0.5)  # l1_ratio balances Lasso (L1) and Ridge (L2)
elastic_net_model.fit(X_train, y_train)

# Predict on training and test sets
elastic_train_pred = elastic_net_model.predict(X_train)
elastic_test_pred = elastic_net_model.predict(X_test)

# Evaluate performance
elastic_train_mse = mean_squared_error(y_train, elastic_train_pred)
elastic_test_mse = mean_squared_error(y_test, elastic_test_pred)
train_ENT_r2 = r2_score(y_train, elastic_train_pred)
test_ENT_r2 = r2_score(y_test, elastic_test_pred)
print(f"ElasticNet Training MSE: {elastic_train_mse}")
print(f"ElasticNet Testing MSE: {elastic_test_mse}")
print(f"ElasticNet Training R2: {elastic_train_mse}")
print(f"ElasticNet Testing R2: {elastic_test_mse}")
```
